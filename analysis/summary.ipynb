# Summary Statistics

```python
import sys
sys.path.insert(0, './load_data')
from load import load_td, load_stats

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### Load Data

```python
stats = load_stats()
td = load_td()
```

### Functions

```python
def getSummaryStats(data):
    mean = np.mean(data)
    median = np.nanmedian(data)
    std = np.std(data)
    iqr = np.nanquantile(data, .75) - np.nanquantile(data, .25)
    minimum = min(data)
    maximum = max(data)
    print('Mean:', mean)
    print('Median:', median)
    print('Standard Deviation:', std)
    print('Interquartile Range:', iqr)
    print('Minimum:', minimum)
    print('Maximum:', maximum)

def plotDistribution(arr, xlabel, bins=150):
    plt.title('Distribution')
    plt.hist(arr, bins=bins)
    plt.xlabel(xlabel)
    plt.ylabel('Frequency')
    plt.minorticks_on()
    plt.xticks(np.arange(0,3000,500))
    plt.grid(axis='x', which='both')
    plt.show()
```

### TD File (all players)

```python
getSummaryStats(td['Rating'])
```

Notice a small spike at every 100 rating

```python
plotDistribution(td['Rating'], 'Final Rating')
```

### Current players

```python
getSummaryStats(stats['Max Rating'])
```

```python
plotDistribution(stats['Max Rating'], 'max ratings')
```
